```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

#### Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

#### Its have two dataset 1 to 121 is bejaia region dataset and from 122 sidi bel abbes region dataset
 we label from 1 to 121 as label 0 
 
 we label from 122 to end as label 1

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
## removing null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]  ## its present as duplicate of head
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns ## column head have some extra spaces
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.head()
```

## Converting required columns to integer type

```python
## till Ws 0 to 5 have integer
df[['day','month','Temperature','RH','Ws','year']]=df[['day','month','Temperature','RH','Ws','year']].astype(int)
```

```python
df.info()
```

```python
[features for features in df.columns if df[features].dtype=='O']
```

```python
Object=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in Object:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## save
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```
